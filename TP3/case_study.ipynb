```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd 
import numpy as np 
```

```python
gdf=gpd.read_file('files\huilerie_sb32_SpatialJoin.shp')
```

```python
gdf.sample(15)
```

```python
gdf['DELEG'].value_counts()
```

```python
fig=plt.figure(figsize=(12,6))
plt.title(' La capacité de stockage total pour chaque délégation', size =20  )
plt.bar(gdf['DELEG'],gdf['CAP_STOCK'])
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```
